```python
import matplotlib.pyplot as plt
import math
x = list(range(201))
for i in range(len(x)):
    x[i] = i*2*math.pi/200
y = []
for i in x:
    calculo = math.sin(i) + 1
    y.append(calculo)
print(len(y), len(x), y)
plt.plot(x, y)
```

```python
x  = [i*4*math.pi/100 for i in range(101)]
y  = [math.sin(i) for i in x]
y1 = [math.cos(i) for i in x]
plt.figure(figsize=(18,3))
plt.plot(x,y,'+-g')
plt.plot(x,y1,'^-y')
plt.title('Función seno $y = \sum \sin(x)$')
plt.xlabel('eje X (metros)')
plt.ylabel('eje Y (kilogramos)')
plt.grid()
plt.text(4, 0.25, 'Jorge 2023')
plt.legend(    ['Seno(x)', 'coseno(x)'] )
```

```python
x  = [i*4*math.pi/40 for i in range(41)]
y  = [math.sin(i)+2 for i in x]
plt.figure(figsize=(18,3))
plt.stem(x,y)
for i,j in zip(x,y):  # acceso por elemento empaquetando x, y
    plt.text(i,j/2,round(j,1), ha='center')
for i in range(len(x)): # acceso por índice
    plt.text(x[i], y[i]+0.1, round(y[i],2), ha='center')
```
